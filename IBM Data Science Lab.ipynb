### My code imports to graph mapping libraries introduced in Week 3, to plot an array of 10 digits, including 0, and negative values

````python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
````

````python
# the array elements
data = np.array([2,3,5,6,28,1,9,-7,4,0])
````

````python
#RUN THIS CELL TO PLOT GRAPH
#create the dataframe for the object, using the subplots method
#plot the data,by passing our data varible which contains the values of the array, to the method
# set the x-y axis with lables and name 
fig, ax = plt.subplots()
ax.plot(data)
ax.set_xlabel('Element Index')
ax.set_ylabel('Values')
ax.set_title('Elements of an Array Plot')
plt.show()
````
